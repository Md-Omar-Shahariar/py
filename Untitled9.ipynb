```python
import matplotlib.pyplot as plt

import numpy as np

import math
```

```python
Trial_list=[100, 1000, 5000, 10000]

for n in Trial_list:
  hits=0
  x_hits=[]
  y_hits=[]
  x_miss=[]
  y_miss=[]
  for i in range(n):
    x  = np.random.uniform(0, 8)
    y = np.random.uniform(0, 4) 
    y1 = math.sqrt(4*x)
    y2 = 8 - x
    if x <= 4:
      if y <= y1:
        hits = hits+1
        x_hits.append(x)
        y_hits.append(y)
      else:
        x_miss.append(x)
        y_miss.append(y)
    else:
      if y <= y2:
        hits = hits+1
        x_hits.append(x)
        y_hits.append(y)
      else:
        x_miss.append(x)
        y_miss.append(y)

  

    
  print(hits)
  area = 56*(hits/n)

  myorder = "For Number of trial  :  {0} and area is : {1}"

  print(myorder.format(n, area))

  plt.scatter(x_hits,y_hits,color ="red",label= "Hit Points")
  plt.scatter(x_miss,y_miss,color ="green",label= "Miss Points")
  plt.legend()
  plt.show()
```
